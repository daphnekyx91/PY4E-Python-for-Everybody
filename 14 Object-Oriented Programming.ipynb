```python
#14.6 Our First Python Object

#Object can contain a number of methods & data items(attributes)

#class keyword contain the data(attribute) and code(method) for the objecy

#class keyword has name of class

class PartyAnimal:       #class keyword acts like a template
    x = 0
    
    def party(self):     #method
        self.x = self.x + 1
        print('So Far', self.x )
        
an = PartyAnimal()        #assign object to class PartyAnimal()

an.party()     #the value of object an will change based on method
an.party()     #the method is within this class so its applicable
an.party()  
PartyAnimal.party(an)
```

```python
#14.7 Classes as types

#in python all variables have a type
#use a built-in dir function to find capabilities of a variable
#use type and dir to find out

class PartyAnimal:
    x =0
    
    def party(self):
        self.x = self.x +1 
        print('So far', self.x)
        
an = PartyAnimal()
print('Type', type(an))
print('Dir', dir(an))
print('Type', type(an.x))
print('Type', type(an.party))
```

```python
#14.8 Object Lifecycle

#defining a class(template)
#when the program finishes, all variables are discarded

class PartyAnimal:
    x = 0
    
    def __init__(self):
        print('I am constructed')
        
    def party(self):
        self.x = self.x + 1
        print('so far', self.x)
        
    def __del__(self):
        print('I am destructed', self.x )
        
an = PartyAnimal()        #class will run through attributes too
an.party()                #method
an.party()

an = 42                    #overide the previous object value
print('an contains', an)   
```

```python
#14.9 Many instances

#making multiple objects from our class
#give each object a different initial value

class PartyAnimal:
    x = 0
    name =''      #empty string
    
    def __init__ (self,nam):            #attribute
        self.name = nam
        print(self.name, 'constructed')
    
    def party(self):                   #method
        self.x = self.x + 1
        print(self.name, 'party count', self.x)
        
s = PartyAnimal('Sally')
s.party()
j = PartyAnimal('Jim')
j.party()
s.party()    #method applied to s, when s stored original val
```

```python
#14.10 Inheritance

#create a new class by extending an existing class
#orignal class => parent class, new class => child class

#extend PartyAnimal to CricketFan

class PartyAnimal:
    x = 0
    name = ''
    def __init__ (self, nam):
        self.name = nam
        print(self.name, 'constructed')
        
    def party(self):
        self.x =self.x + 1
        print(self.name, 'party count', self.x)
```

```python
#import PartyAnimal class into new file and extend

from PartyAnimal import party

class CricketFan(PartyAnimal):
    points = 0
    
    def six(self):
        self.points = self.points + 6
        self.party()   #come from PartyAnimal
        print(self.name, 'points', self.points)

s = PartyAnimal('Sally')
s.party()
j = CricketFan('Jim')
j.party()
j.six()
print(dir(j))
        
```
